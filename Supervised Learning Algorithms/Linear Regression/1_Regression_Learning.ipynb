**`REGRESSION`**
--

**Regression analysis is a statistical method to model the relationship between a dependent (target) and independent (predictor) variables with one or more independent variables. More specifically, Regression analysis helps us to understand how the value of the dependent variable is changing corresponding to an independent variable when other independent variables are held fixed. It predicts continuous/real values such as temperature, age, salary, price, et**

```python
`Below is the mathematical equation for Linear regression:`
```

**`y= mx+c`**  
Here, 

y = dependent variables (target variables),

x = Independent variables (predictor variables),

m = Slope

c = Intercept

**`Types of Regression`**

Each type has its own importance on different scenarios, but at the core, all the regression methods analyze the effect of the independent variable on dependent variables. Here we are discussing some important types of regression which are given below:

>Linear Regression

>Logistic Regression

>Polynomial Regression

>Support Vector Regression

>Decision Tree Regression
